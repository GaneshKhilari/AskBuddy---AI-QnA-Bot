Prompt Templates -Dynamic Prompts

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate     #Thic give us permission to create Dynamic prompts
from langchain_core.output_parsers import StrOutputParser
```

```python
llm = ChatOpenAI(model="gpt-4o")
out = StrOutputParser()
```

```python
prompts =ChatPromptTemplate.from_messages([
    {"role" :"system","content":"you are a Translator and translate input in {language}"},
    {"role":"user","content":"{query}"}
    ])
prompts
```

```python
final_prompt = prompts.format_messages(language ="Hindi",query ="I love Python and Javascript")
result = llm.invoke(final_prompt)
print(result.content)
```

```python
res = llm.invoke(prompts.invoke({"language":"Hindi","query":"I love you"}))
```

```python
print(res.content)
```

```python
chains = prompts | llm | out
```

```python
res = chains.invoke({"language":"Hinglish","query":"I love LangChain!"})
print(res)
```
